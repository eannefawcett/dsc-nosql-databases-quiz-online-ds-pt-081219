## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance ensures that when a server goes down, all information is still available even if the down server contains that information. This is the rationale behind the creation of RDD. They distribute the data to random servers that act as a backup.






## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Hadoop is a type of database that supports MapReduce functionality. With MapReduce, the same code is ran on different servers that have different representative data. After the code is ran, it is then reduced, or aggregated, in a single output.
